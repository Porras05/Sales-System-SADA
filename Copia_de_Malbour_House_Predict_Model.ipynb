# **Melbour Houses**

**Importar librerias y Database**

```python
import pandas as pd
```

```python
melbourne_df = pd.read_csv('melb_data.csv')
melbourne_df
```

```python
melbourne_df.describe()
```

**Limpieza de datos**

```python
melbourne_df = melbourne_df.dropna(axis=0)
melbourne_df
```

```python
y = melbourne_df['Price'] #La variable a definir en este caso Y, tomarÃ¡ el valor del precio
X = melbourne_df[['Rooms',"Bathroom","Landsize","Lattitude","Longtitude"]] #Variables que podrÃ¡n servir para predecir el precio de las casas
```

```python
from sklearn.tree import DecisionTreeRegressor #No se usa un DecisionTreeClassifier ya que se buscan resultados numÃ©ricos

melbourne_model = DecisionTreeRegressor()

melbourne_model.fit(X,y)
```

Predicciones de las primeras casas

```python
print("Predicciones para las primeras 5 casas:")
print(X.head())
print("El precio real es")
print(y.head())
print("Las predicciones son:")
print(melbourne_model.predict(X.head())) #Se usa el mÃ©todo .predict para sacar predicciones de base
```

# Validacion usando mÃ©tricas

**MAE - Mean Absolute Error**

```python
 from sklearn.metrics import mean_absolute_error

 predicted_home_prices = melbourne_model.predict(X)
 mean_absolute_error(y, predicted_home_prices)
```

**ValidaciÃ³n**

```python
from sklearn.model_selection import train_test_split

train_X, val_X, train_y, val_y = train_test_split(X, y)

melbourne_model = DecisionTreeRegressor()

melbourne_model.fit(train_X, train_y)

val_predictions = melbourne_model.predict(val_X)
print(mean_absolute_error(val_y, val_predictions))
```

```python
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes)
    model.fit(train_X, train_y) #Se entrena el modelo
    preds_val = model.predict(val_X)#Se predicen los modelos de variacion de Val_X
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)
```

Ejemplo con cierta cantidad de hojas

```python
for max_leaf_nodes in [5, 50, 500, 1000, 1500, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Nodos de hojas mÃ¡ximo: %d \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
```

**Random Forest**

```python
from sklearn.ensemble import RandomForestRegressor

forest_model = RandomForestRegressor()
forest_model.fit(train_X, train_y)
melb_preds = forest_model.predict(val_X)
print(mean_absolute_error(val_y, melb_preds))
```
